```python
import matplotlib.pyplot as plt
from statistics import mean
```

```python
v = ""
with open("data/averageTravelTimePerMinuteOfTheDayCarriers.txt","r") as f:
    for l in f:
        v = l
carrier_neighbour = [float(x) for x in v.split(',')]
carrier_neighbour = [carrier_neighbour[i:i + 30] for i in range(0, len(carrier_neighbour), 30)]
carrier_neighbour = [mean(x) for x in carrier_neighbour]
```

```python
v = ""
with open("data/averageTravelTimePerMinuteOfTheDayTrains.txt","r") as f:
    for l in f:
        v = l
train = [float(x) for x in v.split(',')]
train = [train[i:i + 30] for i in range(0, len(train), 30)]
train = [mean(x) for x in train]
```

```python
carrier_exchange = []
with open('data/avg_day_c860_p125000_chain.txt','r') as f:
    for l in f:
        carrier_exchange.append(float(l.rstrip('\n')))
carrier_exchange
```

```python
minutes = list(range(0,1440,30))
```

```python
fig, ax = plt.subplots(dpi=100)
ax.grid(color='black', linestyle='-', linewidth=0.1)
ax.plot(minutes, train, color ='red', linewidth=1, label="trains")
ax.plot(minutes, carrier_neighbour, color ='green', linewidth=1, label="carrier neighbour")
ax.plot(minutes, carrier_exchange, color ='green', linewidth=1, label="carrier exchange")
# ax.vlines(train_reach, 0, max(x), linestyles ="dashed", colors ="red")
# ax.vlines(carrier_reach, 0, max(x), linestyles ="dashed", colors ="green")
ax.set_title("Average commuting time by starting time")
ax.set_xlabel("time(min)")
ax.set_ylabel("average commuting time(s)")
leg = ax.legend()
plt.show()
```
